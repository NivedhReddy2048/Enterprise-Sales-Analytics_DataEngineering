```python
files = [
    "Products",
    "Orders",
    "OrderItems",
    "Payments",
    "Employees"
]

for file in files:

    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(f"abfss://sales-data@nivedhmedallion.dfs.core.windows.net/{file}.csv")
    )

    print(f"\n===== {file} =====")
    df.printSchema()
    display(df.limit(5))
```

```python
tables = {
    "products": "Products.csv",
    "orders": "Orders.csv",
    "orderitems": "OrderItems.csv",
    "payments": "Payments.csv",
    "employees": "Employees.csv"
}

for table_name, file_name in tables.items():

    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(f"abfss://sales-data@nivedhmedallion.dfs.core.windows.net/{file_name}")
    )

    (
        df.write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(f"enterprise_databricks.bronze.{table_name}")
    )

    print(f"Loaded: {table_name}")
```

```python
%sql

CREATE SCHEMA IF NOT EXISTS enterprise_databricks.gold;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.product_sales AS

SELECT
    p.ProductID,
    p.ProductName,
    p.Category,

    SUM(oi.Quantity) AS total_quantity_sold,

    ROUND(
        SUM(oi.Quantity * oi.UnitPrice),
        2
    ) AS total_revenue

FROM enterprise_databricks.silver.orderitems oi

JOIN enterprise_databricks.silver.products p
ON oi.ProductID = p.ProductID

GROUP BY
    p.ProductID,
    p.ProductName,
    p.Category;
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.product_sales
ORDER BY total_revenue DESC
LIMIT 10;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.state_sales AS

SELECT
    c.State,

    COUNT(DISTINCT o.OrderID) AS total_orders,

    ROUND(
        SUM(o.OrderAmount),
        2
    ) AS total_revenue

FROM enterprise_databricks.silver.orders o

JOIN enterprise_databricks.silver.customers c
ON o.CustomerID = c.CustomerID

GROUP BY c.State;
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.state_sales
ORDER BY total_revenue DESC
LIMIT 10;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.employee_sales AS

SELECT
    e.EmployeeID,
    e.EmployeeName,
    e.Department,

    COUNT(o.OrderID) AS total_orders,

    ROUND(
        SUM(o.OrderAmount),
        2
    ) AS total_revenue

FROM enterprise_databricks.silver.orders o

JOIN enterprise_databricks.silver.employees e
ON o.EmployeeID = e.EmployeeID

GROUP BY
    e.EmployeeID,
    e.EmployeeName,
    e.Department;
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.employee_sales
ORDER BY total_revenue DESC
LIMIT 10;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.monthly_sales AS

SELECT
    YEAR(OrderDate) AS sales_year,
    MONTH(OrderDate) AS sales_month,

    COUNT(OrderID) AS total_orders,

    ROUND(
        SUM(OrderAmount),
        2
    ) AS total_revenue

FROM enterprise_databricks.silver.orders

GROUP BY
    YEAR(OrderDate),
    MONTH(OrderDate);
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.monthly_sales
ORDER BY sales_year, sales_month;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.payment_analysis AS

SELECT
    PaymentMethod,

    COUNT(*) AS total_payments,

    ROUND(
        SUM(PaymentAmount),
        2
    ) AS total_amount

FROM enterprise_databricks.silver.payments

GROUP BY PaymentMethod;
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.payment_analysis
ORDER BY total_amount DESC;
```

```python
%sql

CREATE OR REPLACE TABLE enterprise_databricks.gold.executive_kpis AS

SELECT
    (SELECT COUNT(*) FROM enterprise_databricks.silver.customers) AS total_customers,

    (SELECT COUNT(*) FROM enterprise_databricks.silver.orders) AS total_orders,

    (
        SELECT ROUND(SUM(OrderAmount),2)
        FROM enterprise_databricks.silver.orders
    ) AS total_revenue,

    (
        SELECT ROUND(AVG(OrderAmount),2)
        FROM enterprise_databricks.silver.orders
    ) AS avg_order_value;
```

```python
%sql

SELECT *
FROM enterprise_databricks.gold.executive_kpis;
```
